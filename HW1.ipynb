```python
#Voters in florida
import re
fv = open("FloridaVoters.html", "r")
line = fv.readline()
big_list = []
while line is not "":
    matches = re.findall(">[a-zA-Z0-9\,\s\-\.]+</td>", line)
    string = "".join(matches)
    string1 = string.lstrip(">")
    string2 = string1.rstrip("<")
    if len(string2) > 0:
        big_list.append(string2)
    #if len(string2) > 0:
        #print string2
    line = fv.readline()
    
#print(big_list)
    
correct_list = []
for text in big_list:
    text = text[:-5]
    correct_list.append(text)
    
#print(correct_list)

w = 1
while w < len(correct_list) - 1:
    if correct_list.index(correct_list[w]) % 6 != 0:
        correct_list[w] = correct_list[w].replace(",","")
        correct_list[w] = int(correct_list[w])
    w += 1 
    
#print(correct_list)

smaller_list = []
z = 0
while z < len(correct_list) - 1:
    short_tuple = tuple(correct_list[z:z+3])
    smaller_list.append(short_tuple)
    z += 6

#print "this is smaller list = ", smaller_list
        
sorted_list = sorted(smaller_list, key = lambda x: x[2])
#print(sorted_list)

fvtext = open("FloridaVOTERS.txt", "w")
for county in sorted_list:
    for text in county:
        fvtext.write("{}".format(text))
        fvtext.write(" ")
    fvtext.write("\n")

fvtext.close()
fv.close()
```
